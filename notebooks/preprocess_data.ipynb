```python
# load IMDB movies data
df_imdb_movies = pd.read_csv(data_path + 'IMDb_movies.csv', low_memory=False)  

df_imdb_movies.head(2)
```

```python
# extract the title from disney data and find the corresponding movie in IMDB data
df_disney['title'] = df_disney['title'].str.lower()

df_imdb_movies['title'] = df_imdb_movies['title'].str.lower()
```

```python
# now use fuzzy matching to find the closest match
from thefuzz import fuzz
from thefuzz import process
```

```python
# compare the imdb title with disney title and then add the imdb rows to disney data, using fuzzy matching
def get_imdb_data(row):
    title = row['title']
    year = row['year']
    ratio = 0
    match = ''
    for i in range(0, len(df_imdb_movies)):
        if df_imdb_movies.iloc[i]['year'] == year:
            if fuzz.ratio(title, df_imdb_movies.iloc[i]['title']) > ratio:
                ratio = fuzz.ratio(title, df_imdb_movies.iloc[i]['title'])
                match = i
    if ratio > 70:
        return df_imdb_movies.iloc[match]
    else:
        return None
```

```python
# get the imdb data for disney movies
df_disney['imdb_data'] = df_disney.apply(get_imdb_data, axis=1)
```
